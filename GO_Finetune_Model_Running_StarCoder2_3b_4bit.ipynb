```python
from transformers import AutoModelForCausalLM, AutoTokenizer


repo_dir = "finetunedGO_starcoder2"
tokenizer = AutoTokenizer.from_pretrained(repo_dir)
model = AutoModelForCausalLM.from_pretrained(repo_dir,use_safetensors=True)


inputs = tokenizer.encode("write a program in GO to check if a number id even or odd", return_tensors="pt").to("cuda")
outputs = model.generate(inputs, eos_token_id=tokenizer.eos_token_id, max_length=100, num_return_sequences=1)
print(tokenizer.decode(outputs[0]))
```
